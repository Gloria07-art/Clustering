```python
import numpy as np
```

```python
# Sample data
data =np.array([[0.77, 0.3], [0.5, 0.88], [0.69, 0.42], [0.41, 0.49], [0.81, 0.33], [0.11, 0.38], [0.73, 0.39], [0.56, 0.75], [0.55, 0.09], [0.44, 0.55], [0.1, 0.89], [0.45, 0.76], [0.75, 0.35], [0.15, 0.29], [0.25, 0.35], [0.22, 0.33], [0.51, 0.69], [0.59, 0.77], [0.81, 0.32], [0.23, 0.31]])

# Reading inputs from the user
# Convert the list to a NumPy array
# Reshape the array into a 3x2 array
cluster_centers = np.fromiter((float(input()) for _ in range(6)), dtype=float).reshape(3, 2)

error = 0
mse = 0
while True:
    # Compute initial distances between each point and cluster center
    distances = np.linalg.norm(data[:, np.newaxis, :] - cluster_centers, axis=2)

    # Assign points to clusters based on the minimum distance
    clusters = np.argmin(distances, axis=1)

    error = mse
    mse = 0

    new_centers = []
    for i, center in enumerate(cluster_centers):
        points_in_cluster = data[clusters == i]
        if len(points_in_cluster) > 0:
            mse += np.sum(np.linalg.norm(points_in_cluster - center, axis=1) ** 2)
            new_center = np.mean(points_in_cluster, axis=0)
            new_centers.append(new_center)

    new_centers = np.array(new_centers)

    # Check for convergence
    if np.isclose(mse , error , rtol=1e-4):
        print(f'{error:.4f}')
        break

    # Update cluster centers
    cluster_centers = new_centers



# 0.85
# 0.14
# 0.32
# 0.76
# 0.21
# 0.36
# 0.4379

```
